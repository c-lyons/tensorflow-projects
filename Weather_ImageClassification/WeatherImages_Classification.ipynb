# Weather Image Classification

Using CNN to classify weather images found here: https://data.mendeley.com/datasets/4drtyfjtfy/1

Images have 4 classes: Rain, Cloudy, Sunshine and Sunrise

Classifications for each image are derived from the image name

## Loading Images and Resizing

```python
import PIL
import numpy as np
from matplotlib import image
import matplotlib.pyplot as plt
import re
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# load all images in a directory
from os import listdir
from matplotlib import image
import cv2

def loadAndResize(sourceDir, targetDir, shape=(32,32)):
    """Load images from user defined directory and resize to user defined shape and write files to target directory"""
    
    loaded_images = [] # initialising empty list to store loaded and resized images
    for filename in listdir(sourceDir):
        try:
            image = cv2.imread(sourceDir+'/'+filename, cv2.IMREAD_COLOR) # load image as color image and resizing
            img_resized = cv2.resize(image, (32, 32))
            img_label = re.findall('([a-zA-Z]+)(?=\d+)', filename)[0]  # unpacking label from  file name <label>1234.jpeg
            loaded_images.append((img_resized, img_label)) # append resized greyscale and label to list as tuple
            print(f'Loaded {filename} as size {img_resized.shape}')
            cv2.imwrite(targetDir+'/'+'resized_'+filename, img_resized)  # write to target dir
            print(f'Wrote resized_'+filename+f' to target directory {targetDir}\n')
        except:
            print(f'Error processing image {filename}\n')  # catching errors
    print('Job Completed Successfully.')
    return loaded_images
```

```python
images = loadAndResize(sourceDir='weatherImages', targetDir='resizedImages')
# cv2.resize(cv2.imread('weatherImages'+'/'+'cloudy164.jpg', cv2.IMREAD_COLOR), (32,32)).shape
```

Splitting data into training and testing sets

```python
# converting list of tuples into DataFrame to split data
full_data = pd.DataFrame(images, columns=['image', 'label'])
full_data.head()
```

```python
full_data.iloc[0]['image'].shape
```

```python
full_data.label.value_counts()
```

```python
# converting categories to integer labels
labelDict = {'sunrise':0, 'cloudy':1, 'shine':2, 'rain':3}
full_data.label.replace(labelDict, inplace=True)
```

```python
X = np.array(list(full_data.image)) # converting series data to expected array format (n_samples, n_pixels, n_pixels)
y = full_data['label']

# test size is hold out test set @10%. Using validation set of 20% of total/ 22.222% of remainig 90%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)
```

```python
W_grid = 4
L_grid = 4

fig, axes = plt.subplots(L_grid, W_grid, figsize = (25, 25))
axes = axes.ravel()

n_training = len(X_train)

for i in np.arange(0, L_grid * W_grid):
    index = np.random.randint(0, n_training) # pick a random number
    axes[i].imshow(X_train[index])
    axes[i].set_title(list(labelDict.keys())[list(labelDict.values()).index(y_train.iloc[index])])
    axes[i].axis('off')
    
plt.subplots_adjust(hspace = 0.4)
```

## Training Keras Model

```python
# converting to type float32
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
y_train
```

```python
X_train
```

```python
import keras

n_cats = len(labelDict)  # number of categories
y_train = keras.utils.to_categorical(y_train, n_cats)
y_test = keras.utils.to_categorical(y_test, n_cats)
```

```python
# scaling X data by max pixel value of 255
X_train = X_train/255
X_test = X_test/255
```

```python
# use below to reshape if 1D arrays
# X_train = np.reshape(X_train, (len(X_train), 32, 32, 1))
# X_test = np.reshape(X_test, (len(X_test), 32, 32, 1))

# checking shape is in (n, 32, 32, 3) format
print(X_train.shape)
print(X_test.shape)
```

```python
# input_shape = X_train.shape
inputShape = X_train.shape[1:]  # skipping n_samples for shape inpy
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
from keras.regularizers import l1_l2
```

```python
# creating convolutional NN-> 
# 2x conv layers-> max pooling + drop out -> 2x conv layers ->...
# max pooling + drop out -> flattening -> 2x dense NN layers -> softmax

drop_out1 = 0.3
drop_out2 = 0.3
l1_l2_reg = 0.00001

cnn_model = Sequential()
cnn_model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu', input_shape=inputShape, kernel_regularizer=l1_l2(l1_l2_reg)))
cnn_model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(drop_out1))

cnn_model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',  kernel_regularizer=l1_l2(l1_l2_reg) ))
# cnn_model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Dropout(drop_out2))

cnn_model.add(Flatten())

# cnn_model.add(Dense(units=128, activation='relu'))
cnn_model.add(Dense(units=32, activation='relu',  kernel_regularizer=l1_l2(l1_l2_reg)))
cnn_model.add(Dense(units=32, activation='relu',  kernel_regularizer=l1_l2(l1_l2_reg)))

cnn_model.add(Dense(units=4, activation='softmax'))
```

```python
cnn_model.compile(loss='categorical_crossentropy', optimizer = keras.optimizers.RMSprop(lr = 0.00005), metrics = ['accuracy'])
```

```python
cnn_model.summary()
```

```python
history = cnn_model.fit(X_train, y_train, batch_size=20, epochs=750, shuffle=True, validation_split=0.2222)
```

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy: {}'.format(evaluation[1]))
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')

plt.title('Model Loss Progression During Training/Validation')
plt.ylabel('Training and Validation Losses')
plt.xlabel('Epoch Number')
plt.legend()
```

```python
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')

plt.title('Model Accuracy Progression During Training/Validation')
plt.ylabel('Training and Validation Accuracy')
plt.xlabel('Epoch Number')
plt.legend(['Training Accuracy', 'Validation Accuracy'])
```
