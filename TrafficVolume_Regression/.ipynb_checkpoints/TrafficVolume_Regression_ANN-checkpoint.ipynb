**Data Set Information:**

Hourly Interstate 94 Westbound traffic volume for MN DoT ATR station 301, roughly midway between Minneapolis and St Paul, MN. Hourly weather features and holidays included for impacts on traffic volume.


**Attribute Information:**

holiday Categorical US National holidays plus regional holiday, Minnesota State Fair
temp Numeric Average temp in kelvin
rain_1h Numeric Amount in mm of rain that occurred in the hour
snow_1h Numeric Amount in mm of snow that occurred in the hour
clouds_all Numeric Percentage of cloud cover
weather_main Categorical Short textual description of the current weather
weather_description Categorical Longer textual description of the current weather
date_time DateTime Hour of the data collected in local CST time
traffic_volume Numeric Hourly I-94 ATR 301 reported westbound traffic volume

### Data Cleaning

Examining and handling outlier values

```python
# there are only about a dozen values with temperatures at 0F therefore can remove with little impact
traffic_df[traffic_df.temp==0]
```

```python
traffic_df.drop(traffic_df[traffic_df.temp==0].index, inplace=True)
```

```python
# outlier data for heavy rainfall (9831mm rain recorded in 1 hour) - obvious error - removing from dataset
traffic_df[traffic_df.rain_1h==traffic_df.rain_1h.describe()['max']]
```

```python
traffic_df.drop(traffic_df[traffic_df.rain_1h==traffic_df.rain_1h.describe()['max']].index, inplace=True)
```

```python
plt.figure(figsize=(20,20))
sns.pairplot(traffic_df, plot_kws={'alpha': 0.05})
```

### Feature Engineering

Extracting features to enhance model. 

Getting days of week from datetime info and encoding categoricals.

```python
traffic_df[traffic_df.holiday!='None']['holiday'].value_counts()
```

Treating all holidays as equal in terms of traffic volume (assumption)

Encoding to binary where 1 is holday, 0 is not holiday

```python
traffic_df['holiday'] = traffic_df.holiday.apply(lambda x: 1 if x!="None" else 0)
```

```python
traffic_df.holiday.value_counts()
```

Extracting Weekday/ Weekend and week of year information from date_time column

```python
# return 1 if day of week < 5, else 0 (monday = 0, tuesday = 1,..., Sunday = 6)
traffic_df['is_weekday'] = traffic_df.date_time.apply(lambda x: 1 if pd.to_datetime(x).weekday() < 5 else 0)
```

```python
# getting week of year from date_time info to capture any temporal effects
traffic_df['week_of_year'] = traffic_df.date_time.apply(lambda x: pd.to_datetime(x).weekofyear)
```

Getting Time of Day information from date_time column

```python
# getting time of day from date_time info to capture any temporal effects
traffic_df['time'] = traffic_df.date_time.apply(lambda x: pd.to_datetime(x).hour)
```

Examining weather_main categorical values and one hot encoding

```python
traffic_df.weather_main.value_counts()
```

```python
temp_df = pd.get_dummies(traffic_df.weather_main, prefix='weather_')
merged_df = traffic_df.merge(temp_df, on=traffic_df.index, how='inner')
```

```python
# cleaning final meged_df by dropping unwanted cols
merged_df = merged_df.drop(['key_0', 'weather_description', 'weather_main', 'date_time'], axis=1)
```

Printing to CSV in directory

```python
merged_df.to_csv('cleaned_data.csv', header=True, index=False)
```

### Data Pre-Processing and Preparation

Normalizing data for input into model

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

scaler = MinMaxScaler()
```

Splitting data into training set X and validation set y

```python
# Defining X and y for train test spllit and reshaping y to correct shape
X = merged_df.drop(['traffic_volume'], axis=1)
y = merged_df['traffic_volume'].values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=123)
```

Fitting scaler to training data only, and transforming all data from that scaler

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```
